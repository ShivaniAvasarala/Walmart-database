```python
## Chnage unit price data type to numeric
df['unit_price']= df['unit_price'].str.replace('$', '').astype(float)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isna().sum()
```

```python
df[df.isna().any(axis=1)]
## Dont see any pattern in the missing values expect payment method
```

```python
#### Drop NA values
df.dropna(inplace=True)
```

```python
df['total'] = df['unit_price'] * df['quantity']
df.head()
```

```python
df.to_csv('clean_data_walmart.csv', index=False)
```
